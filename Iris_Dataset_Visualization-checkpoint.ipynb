```python
"""
Task 1: Exploring and Visualizing the Iris Dataset
DeveloperHubCorportaion AI Industrial Training — Batch 2026
"""

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ─────────────────────────────────────────────
# 1. LOAD THE DATASET
# ─────────────────────────────────────────────
# load via seaborn (no CSV needed)
iris = sns.load_dataset("iris")



# ─────────────────────────────────────────────
# 2. INSPECT THE DATASET
# ─────────────────────────────────────────────
print("=" * 55)
print("IRIS DATASET — BASIC INSPECTION")
print("=" * 55)

print(f"\nShape (rows, columns): {iris.shape}")
print(f"\nColumn names:\n  {list(iris.columns)}")

print("\nFirst 5 rows (.head()):")
print(iris.head())

print("\n.info() — data types and nulls:")
iris.info()

print("\n.describe() — summary statistics:")
print(iris.describe())

print("\nClass distribution:")
print(iris["species"].value_counts())

# ─────────────────────────────────────────────
# 3. SCATTER PLOT — feature relationships
# ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle("Scatter Plots — Iris Dataset", fontsize=14, fontweight="bold", y=1.02)

palette = {"setosa": "#378ADD", "versicolor": "#1D9E75", "virginica": "#D85A30"}

# Plot 1: Sepal length vs Sepal width
sns.scatterplot(
    data=iris,
    x="sepal_length", y="sepal_width",
    hue="species", palette=palette,
    s=70, alpha=0.8, edgecolor="white", linewidth=0.5,
    ax=axes[0]
)
axes[0].set_title("Sepal Length vs Sepal Width")
axes[0].set_xlabel("Sepal Length (cm)")
axes[0].set_ylabel("Sepal Width (cm)")
axes[0].legend(title="Species")

# Plot 2: Petal length vs Petal width
sns.scatterplot(
    data=iris,
    x="petal_length", y="petal_width",
    hue="species", palette=palette,
    s=70, alpha=0.8, edgecolor="white", linewidth=0.5,
    ax=axes[1]
)
axes[1].set_title("Petal Length vs Petal Width")
axes[1].set_xlabel("Petal Length (cm)")
axes[1].set_ylabel("Petal Width (cm)")
axes[1].legend(title="Species")

plt.tight_layout()
plt.savefig("scatter_plots.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n[Saved] scatter_plots.png")

# ─────────────────────────────────────────────
# 4. HISTOGRAMS — value distributions
# ─────────────────────────────────────────────
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
titles   = ["Sepal Length (cm)", "Sepal Width (cm)", "Petal Length (cm)", "Petal Width (cm)"]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Histograms — Feature Distributions by Species", fontsize=14, fontweight="bold")
axes = axes.flatten()

for i, (feat, title) in enumerate(zip(features, titles)):
    for species, color in palette.items():
        subset = iris[iris["species"] == species][feat]
        axes[i].hist(subset, bins=12, alpha=0.6, color=color, label=species, edgecolor="white")
    axes[i].set_title(title)
    axes[i].set_xlabel("Value (cm)")
    axes[i].set_ylabel("Count")
    axes[i].legend(title="Species", fontsize=8)

plt.tight_layout()
plt.savefig("histograms.png", dpi=150, bbox_inches="tight")
plt.show()
print("[Saved] histograms.png")

# ─────────────────────────────────────────────
# 5. BOX PLOTS — outlier detection
# ─────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Box Plots — Outlier Detection by Species", fontsize=14, fontweight="bold")
axes = axes.flatten()

for i, (feat, title) in enumerate(zip(features, titles)):
    sns.boxplot(
        data=iris,
        x="species", y=feat,
        palette=palette,
        width=0.5,
        flierprops=dict(marker="o", markerfacecolor="red", markersize=5, alpha=0.6),
        ax=axes[i]
    )
    axes[i].set_title(title)
    axes[i].set_xlabel("Species")
    axes[i].set_ylabel("Value (cm)")

plt.tight_layout()
plt.savefig("box_plots.png", dpi=150, bbox_inches="tight")
plt.show()
print("[Saved] box_plots.png")

# ─────────────────────────────────────────────
# 6. BONUS — Pair plot (all features at once)
# ─────────────────────────────────────────────
pair_grid = sns.pairplot(iris, hue="species", palette=palette, diag_kind="kde", plot_kws={"alpha": 0.6})
pair_grid.fig.suptitle("Pair Plot — All Feature Combinations", y=1.02, fontsize=14, fontweight="bold")
plt.savefig("pair_plot.png", dpi=150, bbox_inches="tight")
plt.show()
print("[Saved] pair_plot.png")

print("\n✓ Task 1 complete — all visualizations saved.")
```
